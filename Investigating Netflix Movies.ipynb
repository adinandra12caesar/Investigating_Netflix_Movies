```python
# Import the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the dataset as a DataFrame
netflix = pd.read_csv("netflix_data.csv")
netflix.head()
```

```python
# Subset a DataFrame for type Movie
netflix_type_movie = netflix[netflix['type'] == "Movie"]

# Select only the columns of interest
netflix_type_movie_subset = netflix_type_movie.loc[:, ['title', 'country', 'genre', 'release_year', 'duration']]
netflix_type_movie_subset.iloc[0:5]
```

```python
# Set the figure size
fig = plt.figure(figsize = (15, 10))

# Create a scatter plot of release year and duration
plt.scatter(netflix_type_movie_subset['release_year'], netflix_type_movie_subset['duration'])
plt.title("Netflix Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.show()
```

```python
# Filter for durations shorter than 60 minutes
netflix_short_movie = netflix_type_movie_subset[netflix_type_movie_subset['duration'] < 60]
netflix_short_movie.head(25)
```

```python
# Categorize colors by genre
colors = []
for lab, row in netflix_type_movie_subset.iterrows() :
    if row['genre'] == "Children" :
        colors.append("red")
    elif row['genre'] == "Documentaries" :
        colors.append("blue")
    elif row['genre'] == "Stand-Up" :
        colors.append("green")
    else:
        colors.append("black")
        
# Set the figure style and size
plt.style.use("fivethirtyeight")
fig = plt.figure(figsize = (15, 10))

# Create a scatter plot of release year and duration by color categorize
plt.scatter(netflix_type_movie_subset['release_year'], netflix_type_movie_subset['duration'], color = colors)
plt.title("Movie Duration by Year of Release")
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.show()
```
